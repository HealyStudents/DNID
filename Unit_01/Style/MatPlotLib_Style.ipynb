```python
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read in and clean our dataset
iris = datasets.load_iris(as_frame=True).frame
iris["species"] = iris["target"].map({0: "setosa", 1: "versicolor", 2:"virginica"})
iris = iris.rename(columns={"sepal length (cm)":"sepal length", "sepal width (cm)": "sepal width", "petal length (cm)":"petal length", "petal width (cm)":"petal width"})
iris.drop('target', axis=1, inplace=True)
```

```python
iris.head()
```

```python
#The plot we started with before!

fig, ax = plt.subplots()
#the scatter function makes our figure into a scatterplot
ax.scatter(iris["sepal length"], iris["sepal width"])
#One of many functions to alter the figure's style:
plt.title("Sepal Length vs Sepal Width")
#display the figure to the screen
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(iris["sepal length"], iris["sepal width"])
plt.title("Sepal Length vs Sepal Width")

#There are a whole lot of other options we have to change components of our plot
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

#display the figure to the screen
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(iris["sepal length"], iris["sepal width"])

#we can also change fonts!
font = {'family': 'serif', 'color':  'red', 'weight': 'normal', 'size': 50}

plt.title("Sepal Length vs Sepal Width", fontdict=font)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

## The Style Package

In Matplotlib, there is a package that is really helpful for shorthand style changes to plots... it's called `style`

```python
from matplotlib import style 
plt.style.use('default') 
```

```python
#alter the style of this plot
plt.style.use('ggplot') 
# ^^ if you're familiar with ggplot in r, the plot this creates will look familiar

fig, ax = plt.subplots()
ax.scatter(iris["sepal length"], iris["sepal width"])
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
#alter the style of this plot
plt.style.use('fivethirtyeight') 

fig, ax = plt.subplots()
ax.scatter(iris["sepal length"], iris["sepal width"])
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
#The list of available styles is found here:
plt.style.available
```

```python
#Go back to default style
plt.style.use('default') 
```

```python
ax = sns.scatterplot(data=iris, x="sepal length", y="sepal width", hue="petal width", size="petal length", style="species")
sns.move_legend(ax, "right", bbox_to_anchor=(1.3, 0.5))
```

```python
plt.style.use('fivethirtyeight') 
ax = sns.scatterplot(data=iris, x="sepal length", y="sepal width", hue="petal width", size="petal length", style="species")
sns.move_legend(ax, "right", bbox_to_anchor=(1.5, 0.5))
```

```python
ax = sns.scatterplot(data=iris, x="sepal length", y="sepal width", hue="petal width", size="petal length", style="species")
sns.move_legend(ax, "right", bbox_to_anchor=(1.5, 0.5))
ax.set_title("Iris Dataset Summary")
```

```python
#Seaborn has its own style setting shortcuts! Giving set_theme no parameters gives you the seaborn default style
sns.set_theme()
ax = sns.scatterplot(data=iris, x="sepal length", y="sepal width", hue="petal width", size="petal length", style="species")
sns.move_legend(ax, "right", bbox_to_anchor=(1.3, 0.5))
ax.set_title("Iris Dataset Summary")
```

```python
#The set_theme method can take in a whole lot of parameters. "style" is basically the same as the set_style method in matplotlib. 
# palette is a color palette! Check out the docs on supported color palettes: https://seaborn.pydata.org/tutorial/color_palettes.html
sns.set_theme(style="whitegrid", palette="pastel")
#swapped the hue to a categorical variable to show off the new color palette
ax = sns.scatterplot(data=iris, x="sepal length", y="sepal width", size="petal length", hue="species") 
sns.move_legend(ax, "right", bbox_to_anchor=(1.3, 0.5))
ax.set_title("Iris Dataset Summary")
```

```python
#A more vibrant palette and fewer lines! (passing an rc dictionary can allow us to modify a whole lot of very specific parts of our figure)
params = {"axes.spines.right": False, "axes.spines.top": False, "axes.spines.left": False}
sns.set_theme(style="whitegrid", palette="bright", rc=params)
ax = sns.scatterplot(data=iris, x="sepal length", y="sepal width", size="petal length", hue="species") 
sns.move_legend(ax, "right", bbox_to_anchor=(1.3, 0.5))
ax.set_title("Iris Dataset Summary")
```
