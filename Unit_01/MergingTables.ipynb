# Concat
One way of combining datasets is through concatenation. This is appropriate when two (or more) datasets have the *same* columns. It's about extending one table with another **vertically** -- adding the rows of one to another.

For example, in this dataset, we have two tables that contain the same *kind* of data: `diary` and `diary_continued`. These two tables should really just become one big table that represents a union of them. Here's how we do that!

```python
len(diary)
```

```python
len(diary_continued)
```

```python
#this merges these sets together, resulting in a dataset made up of the 1895 rows from diary and the 35 from diary_continued
full_diary = pd.concat([diary, diary_continued])
len(full_diary)
```

```python
#Notice: the column headings are the same as the two dataframes that were concatenated
full_diary.head()
```

```python
# ALSO notice: the indexes got kind of messed up!
for index, row in full_diary.iterrows():
    print(index)
```

# Join
Sometimes we have two tables that relate to each other but have *different* columns. We need to extend our table **horizontally**, so each row contains data from both of our tables.

For example, take the diary table we were just looking at and the `review` table. For each log (row) in the diary table, there is either 1 or 0 corresponding reviews (as in, not every log has a review and every review corresponds with exactly one log).

If you've dealt with databases before, you might recognize that that sounds like a *join*. In pandas, you can do all of the operations one would expect of a database join using the `merge` function.

```python
# let's look at the structure of our data again
full_diary.head()
```

```python
reviews.head()
```

```python
#It looks like *most* of these columns overlap, but "Letterboxd URI" is a kind of unique primary key that we can join on!
diary_with_reviews = full_diary.merge(reviews, on="Letterboxd URI")
diary_with_reviews.head()
```

```python
# ^^^^ this looks like a mess!
# If we merge and columns have fuplicate names, pandas renames the columns so that we have no duplicate names in the resulting table
# We need to fix this... 

# We could start by dropping down to just the few columns we care about
diary_with_reviews = diary_with_reviews[["Date_x", "Name_x", "Year_x", "Letterboxd URI", "Rating_x", "Rewatch_x", "Tags_x", "Watched Date_x", "Review"]]

# And then rename the columns to remove those "_x"
diary_with_reviews = diary_with_reviews.rename(
    mapper={
        "Date_x": "Date", 
        "Name_x": "Name", 
        "Year_x": "Year", 
        "Rating_x": "Rating", 
        "Rewatch_x": "Rewatch", 
        "Tags_x": "Tags", 
        "Watched Date_x": "Watched Date"
    },
    axis=1
)
```

# Querying

We've dealt with querying a good bit before. Here are some other potentially helpful schemes for querying.

```python
#Let's look at watched:
watched.head()
```

```python
# when we query a dataframe, we make a new dataframe copy that matches our query specifications
watched_from_2010 = watched[watched["Year"] == 2010]
watched_from_2010
```

```python
# this is the SAME query, but a different syntax
watched_from_2010 = watched.query("Year == 2010")
watched_from_2010
```

```python
# if we want to combine conditions, we have to use '&' instead of 'and'
# parentheses are HEAVILY recommended, else we get some unexpected errors in these conditions
watched_in_the_2010s = watched[(watched["Year"] >= 2010) & (watched["Year"] < 2020)]
watched_in_the_2010s
```

```python
#There are a number of string methods we could use in queries
blue_films = watched[watched["Name"].str.contains("Blue")]
blue_films
```

# Alternative Forms of Data

There are a few other common forms of data outside of csv and xlsx that pandas can easily handle. 

```python
# Imagine we have a tsv file named "movie_data.tsv"
# a tsv file is like a csv file, but instead of being delimeted by commas, it's delimeted by tabs
tabular_data = pd.read_csv("movie_data.tsv", sep="\t")
```

```python
# Imagine we have a .dta file (a stata file)
# pandas has a method for that!
stata_df = pd.read_stata("example_data.dta")
```

```python
# For an extremely pythonic option, we could use a pickle file
import pickle

# here we take our watched table and save it as a pickle file
with open('watched_pickle.pyc', 'ab') as pickle_file:
    pickle.dump(watched, pickle_file)

# here we take our saved pickle file and read it as a new dataframe
with open('watched_pickle.pyc', 'rb') as pickle_file:    
    watched_copy = pickle.load(pickle_file)

watched_copy.head()
```
